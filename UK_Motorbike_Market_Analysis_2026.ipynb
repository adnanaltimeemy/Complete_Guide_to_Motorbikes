```python
# uk_motorbike_sales_analysis.ipynb
# Step-by-Step Notebook for UK Motorbike Sales + Profit + Demographics

# 1) Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# 2a) Total UK motorcycle registrations (units) 2024‚Äê2025
sales_data = {
    "Year": [2024, 2025],
    # Estimated totals based on industry tracking
    "Total_Registrations": [115000, 94389]  
}

sales_df = pd.DataFrame(sales_data)
sales_df
```

```python
# 3a) Best-selling motorcycle models data
model_data = {
    "Model": [
        "Yamaha NMAX 125", "Honda PCX 125", "Honda CB125F",
        "Triumph Speed 400", "BMW R1250 GS Adventure TE"
    ],
    "Registrations_2024": [3548, 3168, 2133, 1635, 1601]
}

models_df = pd.DataFrame(model_data)
models_df
```

```python
segments = ["125cc/Commuter", "Midrange Road", "Adventure/Touring", "Premium"]
profit_margins = [0.08, 0.12, 0.10, 0.15]   # simulated gross margin %

margin_df = pd.DataFrame({
    "Segment": segments,
    "Gross_Margin": profit_margins
})

margin_df
```

```python
demographics = {
    "Age_Group": ["<25", "25-34", "35-44", "45-54", "55+"],
    "Percentage": [10, 25, 30, 20, 15],
    "Male_%": [85, 88, 90, 87, 80],
    "Female_%": [15, 12, 10, 13, 20]
}

demo_df = pd.DataFrame(demographics)
demo_df
```

```python
plt.figure(figsize=(7,4))
sns.barplot(x="Year", y="Total_Registrations", data=sales_df, palette="Blues_d")
plt.title("UK Motorcycle Registrations 2024 vs 2025")
plt.ylabel("Units Registered")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x="Segment", y="Gross_Margin", data=margin_df, palette="Greens_d")
plt.title("Simulated Profit Margins by Bike Segment")
plt.ylabel("Gross Margin (%)")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x="Age_Group", y="Percentage", data=demo_df, palette="Oranges_d")
plt.title("Simulated UK Rider Age Distribution")
plt.ylabel("Percentage of Riders")
plt.show()
```

```python
print("Avg Motorcycle Sales (2024-25):", sales_df["Total_Registrations"].mean())
print("Top selling 2024 model:", models_df.loc[models_df["Registrations_2024"].idxmax()]["Model"])
```
